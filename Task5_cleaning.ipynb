```python
import pandas as pd
df=pd.read_csv("Titanic-Dataset.csv")
df.head()
df.isnull().sum()
df.describe()
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
df.isnull().sum()
```

```python
df.duplicated().sum()
df=df.drop_duplicates()
df.dtypes
```

```python
df.to_csv("cleaned_titanic_dataset.csv",index=False)
```
